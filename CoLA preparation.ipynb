# Swe-CoLA

```python
data = pd.read_csv('datasets/swe_cola/datasetDaLAJsplit.csv', index_col=0)
```

```python
data
```

```python
true_labels = data[['original sentence', 'error label', 'split']]
false_labels = data[['corrected sentence', 'error label', 'split']]
```

```python
true_labels = true_labels.rename(columns={'original sentence':'sentence'})
true_labels['label'] = 1
false_labels = false_labels.rename(columns={'corrected sentence':'sentence'})
false_labels['label'] = 0
```

```python
data = pd.concat([true_labels, false_labels], axis=0).sample(frac=1.0, ignore_index=True)
```

```python
data['split'].unique()
```

```python
train = data[data['split'] == 'train']
test = data[data['split'] == 'test']
valid = data[data['split'] == 'valid']
```

```python
train.to_csv('datasets/Swe-CoLA-dataset/prepared/train.csv', index=None)
valid.to_csv('datasets/Swe-CoLA-dataset/prepared/valid.csv', index=None)
test.to_csv('datasets/Swe-CoLA-dataset/prepared/test.csv', index=None)
```
